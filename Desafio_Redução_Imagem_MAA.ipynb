Função para converter um array com uma imagem RGB para escala de cinza
    
    Formula: Y = 0.2989 * R + 0.5870 * G + 0.1140 * B
    
  

```python
def converte_rgb_cinza(img_array):
    # Separa os arrays de cada cor
    r = img_array[:, :, 0]
    g = img_array[:, :, 1]
    b = img_array[:, :, 2]
    # Aplica a formula somando as cores no novo array 2D
    img_cinza = 0.2989 * r + 0.5870 * g + 0.1140 * b
    #Retorna o array 2D convertido para inteiro 8 (0 a 255)
    return img_cinza.astype(np.uint8)
```

Função para converter um array em escala de cinza para preto e branco

Utiliza o parametro threshold para definir a linha de corte entre os tons de cinza


```python
def converte_cinza_preto_branco(img_array, threshold):
    img_pb = np.where(img_array >= threshold, 255, 0)
    return img_pb.astype(np.uint8)
```
